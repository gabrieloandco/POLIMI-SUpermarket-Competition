### Missing Values

```python
unknown_per_columns = trainingdata.isnull().sum()
unknowns = unknown_per_columns[unknown_per_columns>0]
unknowns
```

```python
missing_columns= unknowns.index.values
amountofdata_missing_columns= trainingdata[missing_columns].count()
amountofdata_missing_columns
```

```python
percentageofmissing= 100*unknowns/(amountofdata_missing_columns+unknowns)
percentageofmissing
```

We see that 78% of Max_Gust_SpeedKm_h values are missing, assuming that is Missing completely at random (MCAR) then we could drop that row without any further issue.

DO Correlation

```python
processed_trainingdata = trainingdata.drop('Max_Gust_SpeedKm_h',axis=1)
```

Events also appears as a variable with very high amount of missing values

```python
processed_trainingdata['Events'].describe(include="all")
```

```python
processed_trainingdata['Events'].unique()
```

In this case we can see that missing values have a meaning! Then we could replace the missing values with a variable named None, meaning that no event has happened.

```python
processed_trainingdata['Events']= processed_trainingdata['Events'].fillna(value="None")
```

Now we have four numerical variables left: CloudCover , Max_VisibilityKm, Mean_VisibilityKm, Min_VisibilitykM, we can use algorithm to fill them.

```python
from kpopalg import *
missing_columns = ['CloudCover','Max_VisibilityKm','Mean_VisibilityKm','Min_VisibilitykM']
missing_trainingdata = processed_trainingdata[missing_columns]
labels, centroids, kmeans_missing_trainingdata = kmeans_missing(missing_trainingdata, n_clusters=5, max_iter=10)

kmeans_missing_trainingdata
```
